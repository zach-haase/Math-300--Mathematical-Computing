### Problem 1
Plot the function $f(x)=x^2$ for $x\in [-1,1]$ using `np.linspace()` with 10, 100, 1000 slices. Make sure your plot is as informative as possible.

```python
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
fig, axes = plt.subplots(1,3)
axes[1].set_xlabel('x from -1 to 1')
axes[0].set_ylabel('y from 0 to 1')
for i in range(1, 4):
    x = np.linspace(-1,1,10**i)
    a = x**2
    axes[i-1].scatter(x, a)
fig.suptitle('x squared plotted with 10, 100, and 1000 points')
axes[0].set_title('10 points')
axes[1].set_title('100 points')
axes[2].set_title('1000 points')
```
