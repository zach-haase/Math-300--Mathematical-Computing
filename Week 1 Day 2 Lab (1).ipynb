## Problem 8
Suppose

```python
a + b = 25
a * b = 144
```

Find all integer solutions `(a, b)`.

---


````python

for a in range(25):
    for b in range(25):
        if (a*b==144) & (a+b==25):
            print(a)
            print(b)
for a in range(25):
    b = 25-a
    if a*b == 144:
        print(a)
        print(b)
````
