## Problem 6

Define a function

```python
    palindrome(string)
    
```
that takes in a string and outputs whether or not the string is a palindrome. (A palindrome is a word that is the same forward as backward)

````python
def palindrome(string):
    if string[::] == string[::-1]:
        print("palindrome")
    else:
        print ("not palindrome")
string = ("racecar")
palindrome(string)
string = ("bunny")
palindrome(string)
````
