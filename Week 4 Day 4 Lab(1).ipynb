### Problem 3
Create one dataset with a strong positive correlation and another with a strong negative correlation. Plot them side by side and compare their correlation coefficients.

```python
Name = ['Sam Darnold', 'Lamar Jackson', 'Tom Brady']
dashtimesec = [4.85, 4.34, 5.28]
highspeedmph = [21.85, 21.25, 22.7]
throwdistance = [77, 75, 66]
qbstats = pd.DataFrame({
    'Name': Name,
    '40-Yard Dash Time in seconds': dashtimesec,
    'Highest Career Speed in Game in MPH': highspeedmph,
    'Furthest Career Throw Distance in Yards': throwdistance
})
qbstats
```

```python
dash = qbstats['40-Yard Dash Time in seconds']
speed = qbstats['Highest Career Speed in Game in MPH']
throw = qbstats['Furthest Career Throw Distance in Yards']
plt.scatter(dash, speed)
plt.title('QB 40-Yard Dash Speed vs Highest Career MPH in Game')
plt.xlabel('40-Yard Dash Time in Seconds')
plt.ylabel('Career Speed in MPH')
```

```python
corr1 = np.corrcoef(dash, speed)
corr1
```

```python
plt.scatter(dash, throw)
plt.title('QB 40-Yard Dash Speed vs Furthest Career Throw Distance')
plt.xlabel('40-Yard Dash Time in Seconds')
plt.ylabel('Career Throw in Yards')
```

```python
corr2 = np.corrcoef(dash, throw)
corr2
```
