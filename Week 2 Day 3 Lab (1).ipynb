## Problem 2

1. Write a recursive function `factorial()` that takes in a natural number `n` and computes $n!=n(n-1)(n-2)\cdots3\cdot2\cdot 1$. 
2. Use `%timeit` to determine how fast your function calculates `factorial(150)`. 
3. Use `%timeit` to calculate how fast `math.factorial()` calculates the same quantity.
4. Were there any differences between the two methods?

```python
%timeit math.factorial(150)
%timeit factorial(150)
def factorial(n):
    if n <= 1:
        return 1
    else:
        return n * factorial(n - 1)
```
