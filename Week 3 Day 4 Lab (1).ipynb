```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

contgraph = nx.Graph()
type(contgraph)
contgraph.edges()
contgraph.nodes()
#Big Bang Theory Characters
contgraph.add_node('Sheldon')
contgraph.add_node('Leonard')
contgraph.add_node('Howard')
contgraph.add_node('Raj')
contgraph.add_node('Penny')
contgraph.add_node('Amy')
contgraph.add_node('Bernadette')
#Friendships
contgraph.add_edge('Sheldon', 'Leonard')
contgraph.add_edge('Sheldon', 'Howard')
contgraph.add_edge('Sheldon', 'Raj')
contgraph.add_edge('Leonard', 'Howard')
contgraph.add_edge('Leonard', 'Raj')
contgraph.add_edge('Raj', 'Howard')
contgraph.add_edge('Penny', 'Amy')
contgraph.add_edge('Penny', 'Bernadette')
contgraph.add_edge('Bernadette', 'Amy')
#Relationships
contgraph.add_edge('Penny', 'Leonard')
contgraph.add_edge('Sheldon', 'Amy')
contgraph.add_edge('Howard', 'Bernadette')
nx.draw(contgraph,
       with_labels=True,
       node_size = [730, 650, 645, 680, 660, 640, 590],         
       node_color = ['brown', 'brown', 'brown', 'white', 'yellow', 'brown', 'yellow'],
       width = [5, 3, 3, 3, 3, 5, 5, 5, 3, 7, 7, 7],
       edgecolors = ['white', 'white', 'white', 'black', 'white', 'white', 'white']) #raj has black hair but you couldnt read the text over it so i made his node outlined black with white in the middle
plt.title('The Big Bang Theory Characters',
         fontsize=20)
```

For the dolphin social network:
1. Import the data
2. Draw the network 
3. Use as many `networkx` parameters as possible (that is, in `nx.draw()`, add all the parameters: `node_size`, `node_color`, etc.) such that the network is still presentable and understandable

You can find some information about `nx.draw()` here: https://networkx.org/documentation/stable/reference/generated/networkx.drawing.nx_pylab.draw.html#networkx.drawing.nx_pylab.draw .


```python
dolphincsv = pd.read_csv(r"C:\Users\haase\Downloads\Math 300 Downloads\Dolphin thing\DolphinNetwork.csv")
```

```python
dolphincsv
```

```python
edges = [(dolphincsv['Source'].iloc[i], dolphincsv['Target'].iloc[i]) for i in range(len(dolphincsv))]
dolphin = nx.Graph()
dolphin.add_edges_from(edges)
plt.figure(figsize = (10,10))

nx.draw(dolphin,
       node_size=400,
       with_labels=True)
```
