```python
def squares(maximum):

    container = []
```

What is the complexity? Justify your answer. Write a new, simpler function that does the same as `squares()`. Calculate your function's complexity and justify your answer.

````python
#O(3n) because it has three seperate for-loops each going through n and completing a task for each iteration of n
def squares(x):
    a = 0
    b = 0
    for i in range(x):
        a = i**2
        b += a
    return b
squares(50)
#O(n) because I only have one for loop completing a set of tasks for each iteration of n
````

### Problem 5
Write a recursive function `fibonacci(n)` that computes the $n$-th Fibonacci number. Compute the time complexity of your `fibonacci(n)` function.

````python
def fibonacci(n): 
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(7))
#complexity is O(2 ** n)
````

### Problem 6
Write a recursive function `reverse(string)` that reverses the order of a string. Compute the time complexity of your `reverse()` function

````python
def reverse(string):
    if len(string) <= 1:
        return string
    else:
        return string[-1] + reverse(string[:-1])
print(reverse("string"))
#complexity is O(n ** 2)
````

### Problem 7
Write a recursive function that computes the sum of integers up to input $n$.

````python
def sum(n):
    if n <= 0:
        return 0
    else:
        return n + sum(n - 1)
print(sum(5))
````
